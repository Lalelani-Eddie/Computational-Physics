```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Define constants
a = 0.1/2
Vo = 40
c = 3e8
m = (0.511e6)/(c**2)
h = 197.3/c

# Create an array of epsilon values from 0 to 40
epsilon_values = np.linspace(26.34300, 26.344, 1000)  # Adjust the number of points as needed

# Calculate l and k for each epsilon value
l_values = np.sqrt(2 * m * (Vo - epsilon_values) / (h**2))
k_values = np.sqrt(2 * m * epsilon_values / (h**2))

# Define a tolerance level for considering F_epsilon as approximately zero
tolerance = 1e-6
F_epsilon_values = l_values*np.tan(l_values*a) - k_values
# Find and print the values of epsilon where F_epsilon is approximately zero
for i in range(len(epsilon_values)):
    if abs(F_epsilon_values[i]) < tolerance:
        print(f"Epsilon for intercept: {epsilon_values[i]}")
# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(epsilon_values, F_epsilon_values, label='F_epsilon')
plt.xlabel('Epsilon')
plt.ylabel('F_epsilon')
plt.title('Variation of F_epsilon with Epsilon')
plt.grid(True)

# Show the plot
plt.show()
```
